**4.2 Categorical Column:**

1. For character variables, what are the unique values in the variable?
2. Figure out the mode of such variables.
3. If any date variables, what time period do the observations in these data frames span.



```python
# Extracting categorical columns
cat_cols = df.select_dtypes(include=['object'])

# Displaying unique values, mean, and mode for each categorical column
for col in cat_cols:
    print("\nCategorizing Variable: {col}")
    unique_values = df[col].unique()
    mode_value = df[col].mode().iloc[0]

    # Calculating mode for categorical columns
    print("Unique Values: ", unique_values)
    print("Mode: ", mode_value)
    print("")
```

```python
# Extract integer columns
int_cols = df.select_dtypes(include=['int64'])

# Displaying unique values, mean, and mode for each integer column
for col in int_cols:
    print("\nInteger Variable: {col}")
    unique_values = df[col].unique()
    mean_value = df[col].mean()
    mode_value = df[col].mode().iloc[0]

#printing the values
    print("Unique Values: ", unique_values)
    print("Mode: ", mode_value)
    print("")

#The mode value of 0 indicates that the count of 0 occurs most frequently in the dataset.
```

**5. Visualizing the Data:**

```python
# Setting the style for the Seaborn plot with a white grid background
sns.set(style="whitegrid")

# Creating a countplot to visualize the distribution of suicides across different age groups
chart = sns.countplot(x=df['generation'], width=0.7, edgecolor='black')

# Adding labels with count values to each bar in the countplot
chart.bar_label(chart.containers[0])

# Adding x-axis and y-axis labels to the plot
plt.xlabel('Generation')

# Adding a title to the plot with a bold font and a specified fontsize
plt.title('Univariative analysis of suicide rate among generations', weight='bold', fontsize=14)

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

plt.show()
plt.savefig('Hist.png')
```

```python
# Creating a histogram to visualize the distribution of suicide rates based on GDP per capita
plt.hist(df['gdp_per_capita'], bins=10, edgecolor='black')

# Adding a title to the histogram with a bold font and a specified fontsize
plt.title('Histogram of Suicide Distribution', weight='bold', fontsize=14)

# Adding x-axis and y-axis labels to the histogram
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')  # Adjusted y-axis label to 'Frequency'

# Adding grid lines along the y-axis for better readability
plt.grid(axis='y')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=90)

plt.show()

# Saving the histogram plot as an image file (PNG format)
plt.savefig('Hist.png')
```

```python
df = pd.DataFrame({'year': [1985, 1990, 1995, 2000,  2010,2015]})

#Creating a boxplot for the 'year' column
plt.boxplot(df['year'], showmeans=True, meanline=True)

#Adding title and labels
plt.title('Year Boxplot',weight ='bold', fontsize = 20)
plt.ylabel('Year',weight ='bold', fontsize = 20)

plt.savefig('Boxplot')
plt.show()
```

```python
numericalcolumns = ['suicides_no','population']

# Creating a pair plot for the specified numerical columns
sns.pairplot(df[numericalcolumns], height=4)

# Adding a super title to the pair plot with a specified position and transparency
plt.suptitle('Pair Plotting Numerical Variables in Suicide Dataset', y=1.08, alpha=0.5, weight='bold', fontsize=14)

# Save the plot before displaying
plt.savefig('Hist.png')

# Display the plot
plt.show()
```

```python
# Generating a heatmap for the correlation matrix of the DataFrame
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)

#Adding a title to the heatmap with specified weight and fontsize
plt.title('Correlation Figure', weight='bold', fontsize=14)

# Displaying the heatmap
plt.show()

plt.savefig('Hist.png')
```
